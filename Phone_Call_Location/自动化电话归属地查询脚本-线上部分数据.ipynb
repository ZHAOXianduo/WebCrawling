```python
##导入数据处理库
import numpy as np #数据结构库
import pandas as pd #数据结构库
Online_List = pd.read_excel(r'I:/Articles/线上数据.xlsx')
Offline_List = pd.read_excel(r'I:/Articles/线下数据.xlsx')
Online_List['phone_省份'] = 0
Online_List['phone_城市'] = 0
Online_List['phone_运营'] = 0
Offline_List['phone_省份'] = 0
Offline_List['phone_城市'] = 0
Offline_List['phone_运营'] = 0
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
s = Service(r"C:\Users\NHT\Desktop\chromedriver.exe")
browser = webdriver.Chrome(service=s)
browser.get('https://cx.shouji.360.cn/?number=')
time.sleep(3)
now = 0
for i in range(len(Online_List)):
    print('当前进度：',round((i+now)/len(Online_List),6)*100,'%')
    if len(str(Online_List['患者号码'][i+now])) >= 10 and len(str(Online_List['患者号码'][i+now])) <= 17 and str(Online_List['患者号码'][i+now])[0] == '1':
        print(Online_List['患者号码'][i+now])
        input_ = browser.find_element(By.ID,'number')
        input_.send_keys(str(Online_List['患者号码'][i+now]))
        button = browser.find_element(By.ID,'smt')
        button.click()
        info = browser.find_element(By.ID,'banner')
        location = info.text
        wait = 0
        while location == '':
            button.click()
            wait = wait + 0.1
            time.sleep(0.1)
            info = browser.find_element(By.ID,'banner')
            location = info.text
        print('等待时间',wait,'秒')
        List_Loc = location.split(' ')
        print(List_Loc)
        if len(List_Loc) == 3:
            Online_List['phone_省份'][i+now] = List_Loc[0]
            Online_List['phone_城市'][i+now] = List_Loc[1]
            Online_List['phone_运营'][i+now] = List_Loc[2]
        if len(List_Loc) == 2:
            Online_List['phone_省份'][i+now] = List_Loc[0]
            Online_List['phone_城市'][i+now] = List_Loc[0]
            Online_List['phone_运营'][i+now] = List_Loc[1]
        input_.clear()
        button.click()
        info = browser.find_element(By.ID,'banner')
        location = info.text
        while location != '':
            input_.clear()
            button.click()
            wait = wait + 0.1
            time.sleep(0.1)
            info = browser.find_element(By.ID,'banner')
            location = info.text
        print('等待时间',wait,'秒')
        time.sleep(1)
    else:
        pass
```
